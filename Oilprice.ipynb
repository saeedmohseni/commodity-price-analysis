```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Define tickers for Brent Crude and WTI Crude
tickers = {
    "Brent": "BZ=F",
    "WTI": "CL=F"
}

# Download last 30 days of price data
data = yf.download(list(tickers.values()), period="30d", interval="1d")

# Print available columns for debugging
print("Columns in data:", data.columns)

# Extract Close prices (fallback if 'Adj Close' is not available)
if ('Adj Close' in data.columns.get_level_values(0)):
    price_data = data['Adj Close']
else:
    price_data = data['Close']

# Rename columns
price_data.columns = tickers.keys()
price_data.dropna(inplace=True)

# Plot
plt.figure(figsize=(12, 6))
for col in price_data.columns:
    plt.plot(price_data.index, price_data[col], label=col)
plt.title("Oil Price Trend - Last 30 Days (Brent & WTI)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate 7-day change
weekly_change = (price_data.iloc[-1] - price_data.iloc[-7]) / price_data.iloc[-7] * 100
weekly_change = weekly_change.round(2)

# Print changes
print("7-Day Percentage Change:")
print(weekly_change)
```
